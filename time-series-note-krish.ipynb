```python
# This data is seasonal data 
# Now find is it stationary or not ?
```

```python
#Check is this stationary?

# from statsmodels.tsa.stattools import adfuller
# result = adfuller(df["Thousands of Passengers"])
# print(result)

# # Print results
# print('ADF Statistic:', result[0])
# print('p-value:', result[1])
# print('Critical Values:', result[4])

# # Interpretation
# if result[1] < 0.05:
#     print("Series is stationary")
# else:
#     print("Series is not stationary")



#using same thing in function 

from statsmodels.tsa.stattools import adfuller

def adf_test(series):
    
    result = adfuller(series)
    
    print('ADF Statistic: {}'.format(result[0]))
    print('p-value: {}'.format(result[1]))

    # Interpretation
    if result[1] < 0.05:
        print("Series is stationary")
    else:
        print("Series is not stationary")

```

```python
# Using Difference
df["1st_diff"]=df["Thousands of Passengers"]-df["Thousands of Passengers"].shift(1)
df
```

```python
adf_test(df["1st_diff"].dropna())
```

```python
#2nd Diff
df["2nd_diff"]=df["1st_diff"]-df["1st_diff"].shift(1)
df
```

```python
adf_test(df["2nd_diff"].dropna())
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
#acf
#12 months
#Auto correlatiacf
acf=plot_acf(df["2nd_diff"].dropna())
```

```python
pcf=plot_pacf(df["2nd_diff"].dropna())
```

```python
#12 month

df["12_diff"]=df["Thousands of Passengers"]-df["Thousands of Passengers"].shift(12)
df
```
